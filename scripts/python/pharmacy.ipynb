```python
# Let's use SQLite and the sqlite3 library
import sqlite3
import pandas as pd
```

```python
# Viewing the results in the form of DataFrame
def should_display_table_result(db):
  db = cursor_db.fetchall()
  db = pd.DataFrame(db)
  display(db)
```

```python
# Connecting to the database
con_db = sqlite3.connect("pharmacy.db")

# Creating a cursor
cursor_db = con_db.cursor()
```

```python
# Create my table
cursor_db.execute("""
  CREATE TABLE products( 
	  id INTEGER PRIMARY KEY AUTOINCREMENT,
    product_name VARCHAR(255),
    product_type VARCHAR(255),
    retention_revenue BOOLEAN,
    generic_medicine BOOLEAN,
    product_value DECIMAL (8,2)
  )
""")

cursor_db.execute("""
  CREATE TABLE category (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    client_name VARCHAR(255),
    medicine BOOLEAN,
	  cosmetics BOOLEAN,
    personal_hygiene BOOLEAN,

    product_id INTEGER,
    
    foreign key (product_id) references products(id)
  )
""")

# Saving changes
con_db.commit()
```

```python
# Adding my records
product_datas = [
  ['Doril - pct c/ 4 comprimidos', 'analgesico', False, False, 14.95],
  ['Tylenol - 500mg', 'analgesico', False, False, 22.72],
  ['Clonazepam - 2,0mg', 'ansiolitico', True, True, 21.94],
  ['Sabonete Liquido Dove - Nutricao Profunda', 'Outros', False, False, 10.49],
  ['Fralda descartavel Huggies G - 60un', 'Outros', False, False, 57.97],
  ['Creme Facial Eucerin com Ac. Hialuronico + Vit C - 8ml', 'Outros', False, False, 87.92],
  ['Creme dental Sensodyne 100g', 'Outros', False, False, 16.50],
  ['Fluconazol 150mg', 'Antifungico', False, True, 6.43],
  ['Amoxilina 500mg - 21 capsulas', 'Antibiotico', True, True, 15.24],
  ['Velija 60mg', 'Antidepressivo', True, True, 165.50]
]
for data in product_datas:
  cursor_db.execute("INSERT INTO products (product_name, product_type, retention_revenue, generic_medicine, product_value) VALUES (?, ?, ?, ?, ?);", (data[0], data[1], data[2], data[3], data[4]))
  # Saving changes
  con_db.commit()
```

```python
# Fetching records from this table
cursor_db.execute('SELECT * FROM products')

should_display_table_result(cursor_db)
```
